**Question 1**

Given a linked list of **N** nodes such that it may contain a loop.

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**
```
Input:
N = 3
value[] = {1,3,4}
X = 2
Output:1
Explanation:The link list looks like
1 -> 3 -> 4
     ^    |
     |____|
A loop is present. If you remove it
successfully, the answer will be 1.

```
**Example 2:**
```
Input:
N = 4
value[] = {1,8,3,4}
X = 0
Output:1
Explanation:The Linked list does not
contains any loop.

```
**Example 3:**
```
Input:
N = 4
value[] = {1,2,3,4}
X = 1
Output:1
Explanation:The link list looks like
1 -> 2 -> 3 -> 4
^              |
|______________|
A loop is present.
If you remove it successfully,
the answer will be 1.

```



`Approach`:
1. Initialize two pointers, slow and fast, pointing to the head of the linked list.
2. Traverse the linked list using the slow and fast pointers.
- Move slow one step at a time and fast two steps at a time.
- If the fast pointer becomes None, there is no loop in the linked list. Return from the function.
- If the slow and fast pointers meet at some point, there is a loop in the linked list.
3. Reset either the slow or fast pointer to the head of the linked list.
4. Move both pointers one step at a time until they meet again. This meeting point is the start of the loop.
5. Move the fast pointer to the node just before the start of the loop while keeping track of the previous node.
6. Set the next pointer of the previous node to None, breaking the loop and removing the link between the last node and the node at position X.

**Time Complexity**: `O(n)`

**Space Complexity**: `O(1)`

````python
# Definition of the Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to create a linked list from the given values
def createLinkedList(values):
    head = None
    prev = None
    for value in values:
        node = Node(value)
        if head is None:
            head = node
        if prev is not None:
            prev.next = node
        prev = node
    return head

# Function to print the linked list values
def printLinkedList(head):
    curr = head
    while curr is not None:
        print(curr.data, end=" ")
        curr = curr.next
    print()

# Function to remove the loop from the linked list
def removeLoop(head):
    if head is None or head.next is None:
        return head

    slow = head
    fast = head

    # Detect loop
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break

    # If loop is found
    if slow == fast:
        slow = head
        while slow.next != fast.next:
            slow = slow.next
            fast = fast.next
        # Remove the loop
        fast.next = None

    return head

# Given input values
N = 3
values = [1, 3, 4]
X = 2

# Create the linked list
head = createLinkedList(values)

# Create the loop by connecting the last node to the node at position X
if X > 0:
    curr = head
    for _ in range(X - 1):
        curr = curr.next
    last_node = curr
    while curr.next is not None:
        curr = curr.next
    curr.next = last_node

# Remove the loop
head = removeLoop(head)

# Print the updated linked list
printLinkedList(head)
````

**Question 2**

A number **N** is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

**Example 1:**
```
Input:
LinkedList: 4->5->6
Output:457
```

**Example 2:**
```
Input:
LinkedList: 1->2->3
Output:124
```


`Approach`:
1. Reverse the linked list: Reverse the given linked list so that the least significant digit becomes the head of the linked list.
2. Traverse the linked list and add 1: Start from the head and traverse the linked list. Add 1 to the first node (the least significant digit) and propagate the carry if necessary. Keep updating the node values with the sum modulo 10 and set the carry as the sum divided by 10.
3. Handle the carry: After traversing the linked list, if there is still a carry remaining, create a new node with the carry value and append it to the linked list.
4. Reverse the linked list again: Reverse the modified linked list to restore the original order.
5. Return the updated linked list as the result.

**Time Complexity**: `O(n)`

**Space Complexity**: `O(1)`

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def addOne(head):
    # Reverse the linked list
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    head = prev

    # Add 1 to the number
    carry = 1
    prev = None
    curr = head
    while curr:
        sum_val = curr.data + carry
        carry = sum_val // 10
        curr.data = sum_val % 10
        prev = curr
        curr = curr.next

    # If there is a carry after adding 1, create a new node
    if carry > 0:
        prev.next = Node(carry)

    # Reverse the linked list again to get the final result
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    head = prev

    return head

# Function to print the linked list
def printLinkedList(head):
    curr = head
    while curr:
        print(curr.data, end=" -> ")
        curr = curr.next
    print("None")

# Create a sample linked list
def createLinkedList(arr):
    head = None
    prev = None
    for num in arr:
        curr = Node(num)
        if head is None:
            head = curr
        if prev:
            prev.next = curr
        prev = curr
    return head

# Test the implementation with the provided examples
values = [[4, 5, 6], [1, 2, 3]]

for i in range(len(values)):
    head = createLinkedList(values[i])
    print("Original Linked List:")
    printLinkedList(head)

    head = addOne(head)
    print("Linked List after adding 1:")
    printLinkedList(head)
    print()
````

**Question 3**

Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) a **next** pointer to the next node,(ii) a **bottom** pointer to a linked list where this node is head.Each of the sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order. **Note:** The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**

```
Input:
5 -> 10 -> 19 -> 28
|     |     |     |
7     20    22   35
|           |     |
8          50    40
|                 |
30               45
Output: 5-> 7-> 8- > 10 -> 19-> 20->
22-> 28-> 30-> 35-> 40-> 45-> 50.
Explanation:
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

```
**Example 2:**
```
Input:
5 -> 10 -> 19 -> 28
|          |
7          22
|          |
8          50
|
30
Output: 5->7->8->10->19->22->28->30->50
Explanation:
The resultant linked lists has every
node in a single level.

(Note:| represents the bottom pointer.)

```


`Approach`:
1. Traverse the main linked list: Start from the head of the main linked list and traverse each node.
2. Find the tail of the sub-linked list: For each node, traverse the sub-linked list (using the bottom pointer) until you reach the last node (tail) of the sub-linked list.
3. Connect the tail to the next node: Once you find the tail of the sub-linked list, connect it to the next node in the main linked list.
4. Repeat steps 2 and 3 until you reach the end of the main linked list.
5. Return the updated linked list.

**Time Complexity**: `O(n*m)`

**Space Complexity**: `O(1)`

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.bottom = None

def mergeLists(list1, list2):
    # Create a dummy node to store the merged list
    dummy = Node(0)
    tail = dummy

    # Merge the two lists in sorted order
    while list1 and list2:
        if list1.data < list2.data:
            tail.bottom = list1
            list1 = list1.bottom
        else:
            tail.bottom = list2
            list2 = list2.bottom
        tail = tail.bottom

    # Append the remaining nodes of the non-empty list
    if list1:
        tail.bottom = list1
    else:
        tail.bottom = list2

    return dummy.bottom

def flatten(head):
    if head is None or head.next is None:
        return head

    # Recursively flatten the rest of the list
    head.next = flatten(head.next)

    # Merge the current list with the flattened list
    head = mergeLists(head, head.next)

    return head

# Function to print the flattened linked list
def printFlattenedList(head):
    curr = head
    while curr:
        print(curr.data, end=" -> ")
        curr = curr.bottom
    print("None")

# Create a sample linked list
def createLinkedList(arr):
    head = None
    tail = None
    for sublist in arr:
        new_node = Node(sublist[0])
        if head is None:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = tail.next

        for i in range(1, len(sublist)):
            new_node.bottom = Node(sublist[i])
            new_node = new_node.bottom

    return head

# Test the implementation with the provided examples
values = [[[5, 7, 8, 30], [10, 20], [19, 22, 50], [28, 35, 40, 45]],
          [[5, 7, 8, 30], [10, 19], [22, 50]]]

for i in range(len(values)):
    head = createLinkedList(values[i])
    print("Original Linked List:")
    printFlattenedList(head)

    head = flatten(head)
    print("Flattened Linked List:")
    printFlattenedList(head)
    print()
````

**Question 4**

You are given a special linked list with **N** nodes where each node has a next pointer pointing to its next node. You are also given **M** random pointers, where you will be given **M** number of pairs denoting two nodes **a** and **b**  **i.e. a->arb = b** (arb is pointer to random node)**.**

Construct a copy of the given list. The copy should consist of exactly **N** new nodes, where each new node has its value set to the value of its corresponding original node. Both the next and random pointer of the new nodes should point to new nodes in the copied list such that the pointers in the original list and copied list represent the same list state. None of the pointers in the new list should point to nodes in the original list.

For example, if there are two nodes **X** and **Y** in the original list, where **X.arb** **-->** **Y**, then for the corresponding two nodes **x** and **y** in the copied list, **x.arb --> y.**

Return the head of the copied linked list.
**Example 1:**
```
Input:
N = 4, M = 2
value = {1,2,3,4}
pairs = {{1,2},{2,4}}
Output:1
Explanation:In this test case, there
are 4 nodes in linked list.  Among these
4 nodes,  2 nodes have arbitrary pointer
set, rest two nodes have arbitrary pointer
as NULL. Second line tells us the value
of four nodes. The third line gives the
information about arbitrary pointers.
The first node arbitrary pointer is set to
node 2.  The second node arbitrary pointer
is set to node 4.

```
**Example 1:**
```
Input:
N = 4, M = 2
value[] = {1,3,5,9}
pairs[] = {{1,1},{3,4}}
Output:1
Explanation:In the given testcase ,
applying the method as stated in the
above example, the output will be 1.

```

`Approach`:
1. Create a new node for each existing node and insert it next to the original node.
2. Set the random pointers of the new nodes to the corresponding new nodes based on the original random pointers.
3. Separate the original and copied linked lists by restoring the next pointers of the original nodes and extracting the copied nodes.
4. Return the head of the copied linked list.

**Time Complexity**: `O(n)`

**Space Complexity**: `O(1)`

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.random = None

def copyRandomList(head):
    if not head:
        return None
    
    # Step 1: Create a new copy of each node and insert it after the original node
    current = head
    while current:
        new_node = Node(current.data)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next
    
    # Step 2: Set the random pointers of the new nodes
    current = head
    while current:
        if current.random:
            current.next.random = current.random.next
        current = current.next.next
    
    # Step 3: Separate the original and copied linked lists
    current = head
    new_head = head.next
    new_current = new_head
    
    while current:
        current.next = current.next.next
        if new_current.next:
            new_current.next = new_current.next.next
        current = current.next
        new_current = new_current.next
    
    return new_head


# Create the original linked list
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

node1.next = node2
node2.next = node3
node3.next = node4

# Set the random pointers
node1.random = node2
node2.random = node4

# Copy the linked list
copied_head = copyRandomList(node1)

# Print the values of the copied linked list
current = copied_head
while current:
    print(current.data, end=" ")
    current = current.next
````

**Question 5**

Given the `head` of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and return *the reordered list*.

The **first** node is considered **odd**, and the **second** node is **even**, and so on.

Note that the relative order inside both the even and odd groups should remain as it was in the input.

You must solve the problem in `O(1)` extra space complexity and `O(n)` time complexity.

**Example 1:**
```
Input: head = [1,2,3,4,5]
Output: [1,3,5,2,4]

```
**Example 2:**
```
Input: head = [2,1,3,5,6,4,7]
Output: [2,3,6,7,1,5,4]

```



`Approach`:
1. If the linked list is empty or has only one node, return the head as it is.
2. Initialize two pointers, oddHead and evenHead, to keep track of the heads of the odd and even sublists.
3. Create two separate linked lists for odd-indexed and even-indexed nodes by traversing the original linked list.
- Traverse the original linked list using a pointer curr and keep track of the current index.
- For each node, check if the index is odd or even.
- If the index is odd, add the node to the odd sublist using the oddHead pointer.
- If the index is even, add the node to the even sublist using the evenHead pointer.
4. Connect the last node of the odd sublist to the head of the even sublist.
5. Return the head of the odd sublist.

**Time Complexity**: `O(n)`

**Space Complexity**: `O(1)`

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def oddEvenList(head):
    if head is None or head.next is None:
        return head

    # Initialize odd and even sublist heads
    oddHead = head
    evenHead = head.next

    # Initialize odd and even sublist tails
    oddTail = oddHead
    evenTail = evenHead

    # Traverse the original linked list
    curr = evenHead.next
    index = 3

    while curr:
        if index % 2 == 1:
            # Add the node to the odd sublist
            oddTail.next = curr
            oddTail = oddTail.next
        else:
            # Add the node to the even sublist
            evenTail.next = curr
            evenTail = evenTail.next

        curr = curr.next
        index += 1

    # Connect the last node of the odd sublist to the head of the even sublist
    oddTail.next = evenHead

    # Set the next of the last node of the even sublist to None
    evenTail.next = None

    return oddHead

# Function to convert a list to a linked list
def createLinkedList(lst):
    head = ListNode(lst[0])
    curr = head
    for i in range(1, len(lst)):
        curr.next = ListNode(lst[i])
        curr = curr.next
    return head

# Function to convert a linked list to a list
def convertLinkedListToList(head):
    lst = []
    curr = head
    while curr:
        lst.append(curr.val)
        curr = curr.next
    return lst

# Test the implementation with the provided examples
inputs = [[1, 2, 3, 4, 5], [2, 1, 3, 5, 6, 4, 7]]
outputs = [[1, 3, 5, 2, 4], [2, 3, 6, 7, 1, 5, 4]]

for i in range(len(inputs)):
    head = createLinkedList(inputs[i])
    reordered = oddEvenList(head)
    result = convertLinkedListToList(reordered)
    print("Input:", inputs[i])
    print("Output:", result)
    print("Expected Output:", outputs[i])
    print()
````

 **Question 6**

Given a singly linked list of size **N**. The task is to **left-shift** the linked list by **k** nodes, where **k** is a given positive integer smaller than or equal to length of the linked list.

**Example 1:**
```

Input:
N = 5
value[] = {2, 4, 7, 8, 9}
k = 3
Output:8 9 2 4 7
Explanation:Rotate 1:4 -> 7 -> 8 -> 9 -> 2
Rotate 2: 7 -> 8 -> 9 -> 2 -> 4
Rotate 3: 8 -> 9 -> 2 -> 4 -> 7
```
**Example 2:**
```
Input:
N = 8
value[] = {1, 2, 3, 4, 5, 6, 7, 8}
k = 4
Output:5 6 7 8 1 2 3 4

```



`Approach`:
1. Find the kth node from the beginning of the linked list. This will be the new head of the rotated list.
2. If the kth node is not found, return the original list as it is.
3. Find the last node of the original linked list and connect it to the head of the original list to make it a circular linked list.
4. Set the next node of the (k-1)th node as None to break the circularity and create a new tail.
5. Return the new head of the rotated list.

**Time Complexity**: `O(n)`

**Space Complexity**: `O(1)`

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def leftRotate(head, k):
    if head is None or head.next is None or k == 0:
        return head

    # Find the length of the linked list
    length = 1
    curr = head
    while curr.next:
        curr = curr.next
        length += 1

    # Calculate the actual rotation count
    k = k % length

    if k == 0:
        return head

    # Find the kth node from the beginning
    curr = head
    for _ in range(k - 1):
        curr = curr.next

    # Set the new head and new tail
    newHead = curr.next
    newTail = curr

    # Find the last node of the original linked list
    while curr.next:
        curr = curr.next

    # Connect the last node to the head to make it a circular linked list
    curr.next = head

    # Set the next node of the (k-1)th node as None to break the circularity
    newTail.next = None

    return newHead

# Function to convert a list to a linked list
def createLinkedList(lst):
    head = ListNode(lst[0])
    curr = head
    for i in range(1, len(lst)):
        curr.next = ListNode(lst[i])
        curr = curr.next
    return head

# Function to convert a linked list to a list
def convertLinkedListToList(head):
    lst = []
    curr = head
    while curr:
        lst.append(curr.val)
        curr = curr.next
    return lst

# Test the implementation with the provided examples
inputs = [
    ([2, 4, 7, 8, 9], 3),
    ([1, 2, 3, 4, 5, 6, 7, 8], 4)
]
outputs = [
    [8, 9, 2, 4, 7],
    [5, 6, 7, 8, 1, 2, 3, 4]
]

for i in range(len(inputs)):
    head = createLinkedList(inputs[i][0])
    k = inputs[i][1]
    rotated = leftRotate(head, k)
    result = convertLinkedListToList(rotated)
    print("Input:", inputs[i][0])
    print("k:", k)
    print("Output:", result)
    print("Expected Output:", outputs[i])
    print()
````

**Question 7**

You are given the `head` of a linked list with `n` nodes.

For each node in the list, find the value of the **next greater node**. That is, for each node, find the value of the first node that is next to it and has a **strictly larger** value than it.

Return an integer array `answer` where `answer[i]` is the value of the next greater node of the `ith` node (**1-indexed**). If the `ith` node does not have a next greater node, set `answer[i] = 0`.

**Example 1:**
```
Input: head = [2,1,5]
Output: [5,5,0]

```
**Example 2:**
```
Input: head = [2,7,4,3,5]
Output: [7,0,5,5,0]

```


`Approach`:
1. Create an empty stack and an empty result array to store the next greater values.
2. Traverse the linked list from right to left. For each node, perform the following steps:
- a. While the stack is not empty and the value at the top of the stack is less than or equal to the current node's value, pop elements from the stack. These popped elements are the next greater nodes for the current node.
- b. If the stack becomes empty, set the next greater value for the current node as 0.
- c. If the stack is not empty, set the next greater value for the current node as the value at the top of the stack.
- d. Push the current node's value onto the stack.
3. Reverse the result array, as we traversed the linked list from right to left.
4. Return the result array.



**Time Complexity**: `O(n)`

**Space Complexity**: `O(1)`

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def nextLargerNodes(head):
    # Convert the linked list to a list
    values = []
    while head:
        values.append(head.val)
        head = head.next

    # Create an empty stack and an empty result array
    stack = []
    result = [0] * len(values)

    # Traverse the list from right to left
    for i in range(len(values) - 1, -1, -1):
        # Pop elements from the stack until a greater element is found
        while stack and stack[-1] <= values[i]:
            stack.pop()

        # If stack is not empty, set the next greater value
        if stack:
            result[i] = stack[-1]

        # Push the current value onto the stack
        stack.append(values[i])

    return result

# Function to convert a list to a linked list
def createLinkedList(lst):
    head = ListNode(lst[0])
    curr = head
    for i in range(1, len(lst)):
        curr.next = ListNode(lst[i])
        curr = curr.next
    return head

# Test the implementation with the provided examples
inputs = [
    [2, 1, 5],
    [2, 7, 4, 3, 5]
]
outputs = [
    [5, 5, 0],
    [7, 0, 5, 5, 0]
]

for i in range(len(inputs)):
    head = createLinkedList(inputs[i])
    result = nextLargerNodes(head)
    print("Input:", inputs[i])
    print("Output:", result)
    print("Expected Output:", outputs[i])
    print()
````

**Question 8**

Given the `head` of a linked list, we repeatedly delete consecutive sequences of nodes that sum to `0` until there are no such sequences.

After doing so, return the head of the final linked list.  You may return any such answer.

(Note that in the examples below, all sequences are serializations of `ListNode` objects.)

**Example 1:**
```
Input: head = [1,2,-3,3,1]
Output: [3,1]
Note: The answer [1,2,1] would also be accepted.

```
**Example 2:**
```

Input: head = [1,2,3,-3,4]
Output: [1,2,4]

```
**Example 3:**
```
Input: head = [1,2,3,-3,-2]
Output: [1]

```

`Approach`:
1. Create a dummy node and set its next pointer to the head of the given linked list. This dummy node will handle the case where the sum of the entire linked list is zero.
2. Traverse the linked list and calculate the cumulative sum as we move forward. For each node, update its value to the cumulative sum.
3. Create an empty hash table to store the cumulative sum values and their corresponding nodes.
4. Traverse the linked list again and check if the cumulative sum is already present in the hash table. If it is present, it means that there is a subsequence of nodes whose sum is zero. Remove these nodes from the linked list by updating the next pointer of the previous node.
5. If the cumulative sum is not present in the hash table, add it along with the corresponding node to the hash table.
6. Return the next pointer of the dummy node, which will be the head of the final linked list.

**Time Complexity**: `O(n)`

**Space Complexity**: `O(n)`

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeZeroSumSublists(head):
    dummy = ListNode(0)
    dummy.next = head
    cumulative_sum = 0
    node_map = {}
    current = dummy

    while current:
        cumulative_sum += current.val

        if cumulative_sum in node_map:
            prev = node_map[cumulative_sum].next
            temp_sum = cumulative_sum + prev.val
            while temp_sum != cumulative_sum:
                del node_map[temp_sum]
                prev = prev.next
                temp_sum += prev.val
            node_map[cumulative_sum].next = current.next
        else:
            node_map[cumulative_sum] = current

        current = current.next

    return dummy.next

# Function to convert a list to a linked list
def createLinkedList(lst):
    head = ListNode(lst[0])
    curr = head
    for i in range(1, len(lst)):
        curr.next = ListNode(lst[i])
        curr = curr.next
    return head

# Function to convert a linked list to a list
def linkedListToList(head):
    lst = []
    while head:
        lst.append(head.val)
        head = head.next
    return lst

# Test the implementation with the provided examples
inputs = [
    [1, 2, -3, 3, 1],
    [1, 2, 3, -3, 4],
    [1, 2, 3, -3, -2]
]
outputs = [
    [3, 1],
    [1, 2, 4],
    [1]
]

for i in range(len(inputs)):
    head = createLinkedList(inputs[i])
    result = removeZeroSumSublists(head)
    result_list = linkedListToList(result)
    print("Input:", inputs[i])
    print("Output:", result_list)
    print("Expected Output:", outputs[i])
    print()
````
