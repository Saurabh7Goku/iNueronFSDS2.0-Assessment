Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.

</aside>

`Approach`:

 - Define a helper function, convertToDLLUtil, that takes a node as input and returns the modified node at the end of the inorder traversal.
 - In convertToDLLUtil:
    - If the current node is null, return null.
    - Recursively convert the left subtree by calling convertToDLLUtil on the left child of the current node.
    - If the previous node (global variable) is null, set the head of the DLL as the current node; otherwise, update the right pointer of the previous node and the left pointer of the current node to establish the DLL connection.
    - Update the previous node to the current node.
    - Recursively convert the right subtree by calling convertToDLLUtil on the right child of the current node.
    - Return the modified node.
 - Initialize the previous node as null.
 - Call the convertToDLLUtil function with the root of the binary tree.
 - The DLL is now created and the head node is the first node of the inorder traversal.

**Time Complexity O(N)**     
**Space Complexity O(N)**

```python
from collections import deque


class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def convertToDLLUtil(root):
    global prev

    if root is None:
        return None
    head = convertToDLLUtil(root.left)

    if prev is None:
        head = root
    else:
        root.left = prev
        prev.right = root
    prev = root

    convertToDLLUtil(root.right)

    return head


def convertToDLL(root):
    global prev
    prev = None

    head = convertToDLLUtil(root)

    return head

def printf(root):

	if (root == None):
		return
	q = deque()
	q.append(root)
	while (1):
		nodeCount = len(q)
		if (nodeCount == 0):
			break
		while (nodeCount > 0):
			node = q.popleft()
			print(node.data, end=' ')

			if (node.left != None):
				q.append(node.left)

			if (node.right != None):
				q.append(node.right)
			nodeCount -= 1

root = Node(10)
root.left = Node(12)
root.right = Node(15)
root.left.left = Node(25)
root.left.right = Node(30)
root.right.left = Node(36)

head = convertToDLL(root)
printf(head)
```

In the flip operation, the leftmost node becomes the root of the flipped tree and its parent becomes its right child and the right sibling becomes its left child and the same should be done for all left most nodes recursively.

</aside>

`Approach`:

 - Define a function, flipBinaryTree, that takes a node as input.
 - If the node is null or a leaf node (no children), return the node.
 - Recursively flip the right subtree by calling flipBinaryTree on the right child of the current node.
 - If the left child of the current node exists:
    - Recursively flip the left subtree by calling flipBinaryTree on the left child.
    - Set the left child's right pointer to the current node.
    - Set the left child's left pointer to the right sibling of the current node (i.e., the right child).
 - Set the current node's left and right pointers to null.
 - Return the node.
 - Call the flipBinaryTree function with the root of the binary tree.

**Time Complexity O(N)**      
**Space Complexity O(N)**

```python
from collections import deque

class Node:

	def __init__(self, key):
	
		self.data = key
		self.left = None
		self.right = None

def flipBinaryTree(root):
	curr = root
	next = None
	temp = None
	prev = None


	while(curr):
		next = curr.left
		curr.left = temp
		temp = curr.right

		curr.right = prev

		prev = curr
		curr = next
	return prev

def printf(node):
    if node is None:
        return
    print(node.data, end=" ")
    printf(node.left)
    printf(node.right)
        

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.right.left = Node(4)
root.right.right = Node(5)

root = flipBinaryTree(root)
printf(root)
```

ðŸ’¡ Question-3:

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

Input:

        6
     /    \
    3      5
  /   \     \
 2     5     4
     /   \
    7     4

Output:

There are 4 leaves, hence 4 root to leaf paths -
  6->3->2
  6->3->5->7
  6->3->5->4
  6->5>4

</aside>

`Approach`: 

 - Define a function, printRootToLeafPaths, that takes the root of the binary tree as input.
 - If the root is null, return.
 - Initialize an empty stack and a dictionary to track the paths from the root to each node.
 - Push the root node and its value as a string into the stack.
 - Push an empty string into the dictionary with the root node's value as the key.
 - Iterate until the stack is empty:
    - Pop a node and its path from the stack.
    - If the node is a leaf node, print the path.
    - If the node has a right child, push the right child and the updated path (including the right child's value) into the stack.
    - If the node has a left child, push the left child and the updated path (including the left child's value) into the stack.
 - Call the printRootToLeafPaths function with the root of the binary tree.

**Time Complexity O(N)**    
**Space Complexity O(N)**    

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def printRTL(root):
    if root is None:
        return

    stack = [(root, str(root.data))]
    paths = {root.data: ''}

    while stack:
        node, path = stack.pop()

        if node.left is None and node.right is None:
            print(path)
        
        if node.right:
            stack.append((node.right, path + '->' + str(node.right.data)))
            paths[node.right.data] = path + '->' + str(node.right.data)

        if node.left:
            stack.append((node.left, path + '->' + str(node.left.data)))
            paths[node.left.data] = path + '->' + str(node.left.data)


root = Node(6)
root.left = Node(3)
root.right = Node(5)
root.left.left = Node(2)
root.left.right = Node(5)
root.right.right = Node(4)
root.left.right.left = Node(7)
root.left.right.right = Node(4)

printRTL(root)
```

Output :

Explanation :

Input : 

        Inorder -> 4 2 5 1 3

Output : 

No

</aside>

`Approach`: 

 - Define a function, checkSameTree, that takes the pre-order, in-order, and post-order traversal lists as input.
 - If the lengths of the three lists are not equal or any of them is empty, return False.
 - If the length of any list is 1, return True since a single node is always the same tree.
 - Select the first element from the pre-order list as the root of the tree.
 - Find the index of the root value in the in-order list.
 - Split the in-order list into left and right subtrees based on the index found in the previous step.
 - Split the pre-order and post-order lists into left and right subtrees using the length of the left subtree from step 6.
 - Recursively call checkSameTree with the left subtree pre-order, in-order, and post-order lists.
 - Recursively call checkSameTree with the right subtree pre-order, in-order, and post-order lists.
 - Return the logical AND of the two recursive calls.
 - Call the checkSameTree function with the given pre-order, in-order, and post-order traversal lists.

**Time Complexity O(N)**    
**Space Complexity O(N)**    

```python
def checkSameTree(preorder, inorder, postorder):
    if len(preorder) != len(inorder) or len(inorder) != len(postorder) or len(preorder) == 0:
        return False

    if len(preorder) == 1:
        return True

    root = preorder[0]
    root_index = inorder.index(root)

    left_inorder = inorder[:root_index]
    right_inorder = inorder[root_index+1:]

    left_preorder = preorder[1:root_index+1]
    right_preorder = preorder[root_index+1:]

    left_postorder = postorder[:root_index]
    right_postorder = postorder[root_index:len(postorder)-1]

    return checkSameTree(left_preorder, left_inorder, left_postorder) and \
           checkSameTree(right_preorder, right_inorder, right_postorder)

inorder = [4, 2, 5, 1, 3]
preorder = [1, 2, 4, 5, 3]
postorder = [4, 5, 2, 3, 1]

print(checkSameTree(preorder, inorder, postorder)) 

Inorder =  [4, 2, 5, 1, 3]
Preorder = [1, 5, 4, 2, 3]
Postorder = [4, 1, 2, 3, 5]

print(checkSameTree(Preorder, Inorder, Postorder))
```
